```python
import pandas as pd
df = pd.read_csv('IPL.csv')
#display first five rows
df.head()
```

```python
#Explore the dataset structure
df.info()
```

```python
#Check missing values for every column
df.isnull().sum()
```

```python
# Check missing values
print(df.isnull().sum())

# Fill numerical missing values with the median
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

for col in numeric_cols:
    df[col] = df[col].fillna(df[col].median())

# Fill text (object) columns with "Unknown"
object_cols = df.select_dtypes(include=['object']).columns

for col in object_cols:
    df[col] = df[col].fillna("Unknown")

# Fill Boolean columns with False
bool_cols = df.select_dtypes(include=['bool']).columns

for col in bool_cols:
    df[col] = df[col].fillna(False)

# Verify that missing values have been handled
print(df.isnull().sum())
```

```python
# Check duplicate records
print("Duplicate rows:", df.duplicated().sum())

# Remove duplicate records
df.drop_duplicates(inplace=True)

# Verify duplicates are removed
print("Duplicate rows after cleaning:", df.duplicated().sum())
```

```python
# Remove extra spaces and convert text to title case
text_cols = df.select_dtypes(include=['object']).columns

for col in text_cols:
    df[col] = df[col].str.strip().str.title()

print("Text columns cleaned successfully.")
```

```python
# Convert date column to datetime format
df["date"] = pd.to_datetime(df["date"], errors="coerce")

# Ensure ball number is float
df["ball_no"] = df["ball_no"].astype(float)

print(df.dtypes)
```

```python
# Rename columns for consistency
df.rename(columns={
    "ball_no": "ball_number",
    "bat_pos": "batting_position",
    "runs_extras": "extra_runs",
    "runs_total": "total_runs",
    "runs_bowler": "bowler_runs",
    "team_wicket": "team_wickets",
    "event_match_no": "event_match_number",
    "match_number": "match_no"
}, inplace=True)
```

```python
 # Remove unnecessary column
df.drop(columns=["Unnamed: 0"], inplace=True)
 # Create Strike Rate
df["strike_rate"] = (df["batter_runs"] / df["batter_balls"]) * 100
```

```python
# Display updated dataset information
print(df.info())

# Display first five rows
print(df.head())

# Check remaining missing values
print(df.isnull().sum())

# Check duplicates
print(df.duplicated().sum())
```

```python
# Save the cleaned dataset
df.to_csv("cleaned_dataset.csv", index=False)

print("Dataset cleaned successfully!")
```
